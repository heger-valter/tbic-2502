```python
#1a
import pandas as pd
dl = pd.read_csv('license.csv' )
#print(dl)

dl_sorted = dl.sort_values(by='ElevatingDevicesLicenseNumber')
print(dl_sorted)
```

```python
def get_last_word(s: str) -> str:
    return s.strip().split()[-1] if s.strip() else ''
    
def applytocol(col):
    col['city'] =  get_last_word(col)
        
applytocol( dl_sorted[ 'LICENSEHOLDERADDRESS'] )
print(dl_sorted)
```
