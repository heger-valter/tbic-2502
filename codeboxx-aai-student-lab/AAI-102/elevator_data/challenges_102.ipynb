```python
#1a
import pandas as pd
dl = pd.read_csv('license.csv' )
#print(dl)

dl_sorted = dl.sort_values(by='ElevatingDevicesLicenseNumber')
print(dl_sorted)
```

```python
import re

def normalize_spaces(s: str) -> str:
    return re.sub(r'\s+', ' ', s).strip()
    
def get_last_word(s: str) -> str:
    return s.strip().split()[-1] if s.strip() else ''
    
def get_nth_from_end(s: str, delimiter: str, n: int) -> str:
    parts = s.split(delimiter)
    return parts[-n] if 0 < n <= len(parts) else ''
    
def applytocountry(df, col_series):
    """
    Loops through each address in col_series (like df['LocationoftheElevatingDevice'])
    and extracts the last word, placing it in df['country'] for that row only.
    """
    for idx, addr in col_series.items():
        if is_string(addr)  :
            # Normalize spaces if you want consistent spacing
            clean_addr = normalize_spaces(addr)
        
            # Extract the last word (country)
            last = get_last_word(clean_addr)
        
            # Store this row's extracted value in df['country'] for the same row index
            if last != "CA":
                print( "NOt CA, "+ addr + "  " + last)# debug
                
            df.loc[idx, 'country'] = last
        else :
            df.loc[idx, 'country'] = ""
        
def is_string(value):
    return isinstance(value, str)
    
def applytoprovince( df, col_series):
    """
    Loops through each address in col_series, extracts the 4th word from the end
    (could be 'ON', 'NY', etc.), and assigns it to df['state'] for each row.
    """
    for idx, addr in col_series.items():
        if is_string(addr)  :
            clean_addr = normalize_spaces(addr)# we did this before so maybe look at the cached version
            #if df.loc[idx, 'country'] != 'CA' :
            #    print()
                # The 4th word from the end might be the state/province in your addresses
            st = get_nth_from_end(clean_addr, ' ', 2)  # maybe this might need to be tweaked for the USA
        
            # Update the row's 'province' column
            df.loc[idx, 'province'] = st
        else :
#            string_idx = "%d" % idx
#            print ( "error idx=" + string_idx + "  " + f"{addr}")
            df.loc[idx, 'province'] = ""


def applytocity( df, col_series):
    """
    Loops through each address in col_series, extracts the 5th word from the end
    (might be 'Toronto', 'London', etc.), and stores it in df['city'].
    """
    for idx, addr in col_series.items():
        if is_string(addr)  :

            clean_addr = normalize_spaces(addr)
        
            # The 5th word from the end could be the city
            city_val = get_nth_from_end(clean_addr, ' ', 5)
        
            # Update the row's 'city' column
            df.loc[idx, 'city'] = city_val
            
        else :
            df.loc[idx, 'city'] = "", 
      
# 1) Apply to 'LICENSEHOLDERADDRESS' (or whichever column truly has your addresses)
applytocountry( dl_sorted, dl_sorted['LocationoftheElevatingDevice'])
applytocity( dl_sorted, dl_sorted['LocationoftheElevatingDevice'])
applytoprovince( dl_sorted, dl_sorted['LocationoftheElevatingDevice'])

# 2) Print the results
#print( dl_sorted[['LocationoftheElevatingDevice', 'country', 'province', 'city']].head(10))

print(dl_sorted)
```
